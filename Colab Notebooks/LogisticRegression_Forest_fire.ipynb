#Importation des bibliotheques

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import warnings
import pickle
import chardet
#metrics
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve, precision_recall_curve
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import confusion_matrix, log_loss
from sklearn.model_selection import learning_curve
```

#Afficher le Dataset Forest fire

```python
warnings.filterwarnings("ignore")

with open('/content/drive/MyDrive/Colab Notebooks/Forest_fire.csv', 'rb') as f:
    result = chardet.detect(f.read())  # or readline if the file is large
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Forest_fire.csv', encoding=result['encoding'])


data.head(10)
```

#Separer les X des y

```python
data = np.array(data)

X = data[1:, 1:-1]
y = data[1:, -1]
y = y.astype('int')
X = X.astype('int')
```

#Split les donnees en Training et Testing

```python
# print(X,y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

#Creation et Entrainnement du modele LogisticRegression

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
y_pred
```

#Afficharge des metrics

```python
print('Precision: %.3f' % precision_score(y_test, y_pred))
print('Recall: %.3f' % recall_score(y_test, y_pred))
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
print('F1 Score: %.3f' % f1_score(y_test, y_pred))
```

```python
inputt=[int(x) for x in "45 32 60".split(' ')]
final=[np.array(inputt)]

b = log_reg.predict_proba(final)
b
```

#Affichage des probabilites du testing

```python
prediction = log_reg.predict_proba(X_test)
prediction
```

```python
df_prediction = pd.DataFrame({'Actual': y_test, 'Predicted':y_pred})
df_prediction
```

#Calculer la fonction de coût (log loss) sur l'ensemble de test

```python
# Calculer la fonction de coût (log loss) sur l'ensemble de test
cost = log_loss(y_test, log_reg.predict_proba(X_test))

# Affichage du résultat
print(f"Fonction de coût (log loss) : {cost}")
```

#Evaluation du modole de regression logistique

```python
#Create classification report
from sklearn.metrics import classification_report

class_report=classification_report(y_test, y_pred)
print(class_report)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

```python
# Extract the true positive, true negative, false positive, and false negative values from the confusion matrix
tn, fp, fn, tp = conf_matrix.ravel()

# Print the true positive, true negative, false positive, and false negative values
print("True Positive (TP): ", tp)
print("True Negative (TN): ", tn)
print("False Positive (FP): ", fp)
print("False Negative (FN): ", fn)
```
