## Fonction qui recupere la cle de la valeur maximale 

```python
def Valeur_Max(df):
    for cle,valeur in df.items():
        if valeur == df.max():
            return cle
```

## Remplir les NaN des features 'Object'

```python
for i in range(0, df.shape[1]):
    col_name = df.columns[i]
    if((df[col_name]).dtypes==object):
        print(df[col_name].value_counts())
        print('--------------------------------------------')
        val_col_name = Valeur_Max(df[col_name].value_counts())
        print(val_col_name)
        df[col_name].fillna(value=val_col_name, inplace=True)
```
