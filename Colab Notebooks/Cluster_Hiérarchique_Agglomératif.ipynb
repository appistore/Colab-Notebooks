# Cas Pratique : Analyse de Cluster Hiérarchique Agglomératif
**Contexte** :<br>
Une entreprise de vente en ligne souhaite améliorer son marketing en comprenant mieux le comportement d'achat de ses clients. Ils ont recueilli des données sur les achats récents de 500 clients, y compris le montant total dépensé et le nombre de produits achetés. <br>

**Objectif** :<br>
Utiliser l'analyse de cluster hiérarchique agglomératif pour regrouper les clients en segments homogènes en fonction de leurs habitudes d'achat.<br>

Étapes : <br>
Collecte des Données :<br>

Montant Total Dépensé (en dollars)<br>
Nombre de Produits Achetés<br><br>
Prétraitement des Données :<br>

Normalisation des données si nécessaire pour que les variables aient la même échelle.<br>
Vérification et gestion des valeurs manquantes.<br><br>

Choix de la Méthode de Similarité :<br>

Utilisation de la distance euclidienne comme mesure de similarité entre les clients.<br><br>
Construction du Dendrogramme :<br>

Utilisation d'un dendrogramme pour visualiser les regroupements à différents niveaux de similarité.<br>
Identification du nombre optimal de clusters en observant la hauteur des branches du dendrogramme.<br><br>
Exécution de l'Algorithme :<br>

Utilisation d'un algorithme de cluster hiérarchique agglomératif pour regrouper les clients en clusters.<br>
Application de la méthode de fusion (single linkage, complete linkage, average linkage, etc.) en fonction des besoins.<br><br>
Interprétation des Clusters :<br>

Analyse des caractéristiques de chaque cluster, y compris les habitudes d'achat spécifiques.<br>
Donner des noms ou des labels aux clusters en fonction des profils des clients.<br><br>

Évaluation et Validation (si nécessaire) :<br>
Utilisation de mesures d'évaluation telles que l'indice de silhouette pour évaluer la qualité des clusters obtenus.<br>
Réajustement des paramètres ou des méthodes si les résultats ne sont pas satisfaisants.<br>

```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram

# Exemple de données (remplacez cela par vos propres données)
data = {
    'MontantDepense': [100, 250, 150, 700, 800, 200, 50, 300, 1000, 900],
    'NombreProduits': [3, 5, 4, 10, 12, 4, 2, 6, 15, 14]
}

df = pd.DataFrame(data)

# Normalisation des données
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Calcul de la matrice de liaison (linkage matrix)
linkage_matrix = linkage(scaled_data, method='ward')  # Vous pouvez choisir une autre méthode si nécessaire

# Création du dendrogramme
dendrogram(linkage_matrix, labels=df.index, orientation='top')
plt.xlabel('Index des clients')
plt.ylabel('Distance Euclidienne')
plt.title('Dendrogramme')
plt.show()

# Modèle de clustering hiérarchique agglomératif
num_clusters =4  # Nombre de clusters à identifier (selon le dendrogramme)
clustering = AgglomerativeClustering(n_clusters=num_clusters)
df['Cluster'] = clustering.fit_predict(scaled_data)

# Affichage des clusters
print('\n')
print(df)
```

**Résultats et Interprétation :**
Après l'exécution de l'algorithme de cluster hiérarchique agglomératif et l'analyse du dendrogramme, 4 clusters distincts ont été identifiés :<br><br>

**Cluster Premium : (3) **

Clients ayant dépensé beaucoup d'argent et acheté un grand nombre de produits.
Peut-être des clients réguliers ou des entreprises effectuant des achats en gros.<br><br>

**Cluster Moyen Premium : (2) **

Clients ayant dépensé un montant d'argent et acheté un  nombre de produits juste au dessus de la moyenne.
Peut-être des clients réguliers ou des entreprises effectuant des achats en gros.<br><br>

**Cluster Standard : (1) **

Clients ayant dépensé un montant modéré et acheté un nombre moyen de produits.
Peut-être des clients occasionnels ou des particuliers effectuant des achats réguliers.<br><br>

**Cluster Économique : (0) **

Clients ayant dépensé peu d'argent et acheté un nombre limité de produits.
Peut-être des clients à budget serré ou des acheteurs occasionnels.
