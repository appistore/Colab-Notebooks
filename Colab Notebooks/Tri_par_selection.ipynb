#Tri par Selection en utilisation une fonction

```python
def tri_par_selection(list):
	for l in range(len(list)):
		min_i = l
		for i in range(l, len(list)):
			if list[i] < list[min_i]:
				min_i = i
		list[l], list[min_i] = list[min_i], list[l]
	return list
```

```python
import numpy as np
L = [np.random.randint(20) for k in range(10)]
#print(L)

print("--------------")
print("Le tri par selection de la liste ", L , " est : " , tri_par_selection(L))
```
