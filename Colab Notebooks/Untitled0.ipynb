```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Exemple de données (remplacez cela par vos propres données)
X = np.array([[1, 2], [5, 8], [1.5, 1.8], [8, 8], [1, 0.6], [9, 11]])


# Création du modèle de clustering hiérarchique avec la méthode de liaison 'ward'
linked = linkage(X, 'ward')

# Affichage du dendrogramme
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', labels=range(1, len(X) + 1), distance_sort='descending', show_leaf_counts=True)
plt.xlabel('Points de Données')
plt.ylabel('Distance Euclidienne')
plt.title('Dendrogramme')
plt.show()

# Création du modèle de clustering hiérarchique divisif avec 2 clusters
n_clusters = 2
model = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
model.fit(X)
#X['Cluster'] = model.fit_predict(X)

# Récupération des étiquettes de cluster pour chaque point
labels = model.labels_
print('\n')
# Affichage des résultats
for i in range(n_clusters):
    print(f'Cluster {i + 1}:')
    cluster_points = X[labels == i]
    print(cluster_points)
    print('---')

print(X)
```
