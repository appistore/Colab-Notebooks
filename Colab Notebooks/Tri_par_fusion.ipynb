# Tri par Fusion

```python
# Pour L et M deux listes triees, renvoie la liste triee qui contientles elemets de  et M
def Fusion(L,M):
	N=[]
	while L!= [] and M!=[]:
		if L[0] < M[0]:
			N.append(L[0])
			L.pop(0)
		else:
			N.append(M[0])
			M.pop(0)
	return N+L+M
```

```python
def TriFusion(L):
	#Tri de la liste L
	if len(L) <= 1: # La liste est donc forcement triee
		return L
	G = L[0:len(L)//2] 		#Partie gauche de L
	D = L[len(L)//2:len(L)] #Partie droite de L

	G = TriFusion(G) # la liste G est maintenent triee
	D = TriFusion(D) # la liste D est maintenent triee

	F = Fusion(G,D) # la liste F esr triee

	return F
```

```python
import numpy as np
L = [np.random.randint(20) for k in range(10)]
#print(L)

L_triee = TriFusion(L)
print("--------------")
print("Le tri par fusion de la liste ", L , " est : " , L_triee)
```
